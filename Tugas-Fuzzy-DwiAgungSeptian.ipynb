```python
#Dwi Agung Septian
#221011400470

import numpy as np

# Fuzzy Membership Functions
def turun(x, a, b):
    if x <= a: return 1
    if x >= b: return 0
    return (b - x) / (b - a)

def naik(x, a, b):
    if x <= a: return 0
    if x >= b: return 1
    return (x - a) / (b - a)

def segitiga(x, a, b, c):
    if x <= a or x >= c: return 0
    if x == b: return 1
    if x < b: return (x - a) / (b - a)
    return (c - x) / (c - b)

# Produksi Bertambah dan Berkurang
def produksi_berkurang(alpha, min_val, max_val):
    return max_val - alpha * (max_val - min_val)

def produksi_bertambah(alpha, min_val, max_val):
    return min_val + alpha * (max_val - min_val)

# FIS Tsukamoto
def fuzzy_inference(permintaan, persediaan):
    # Parameter
    produksi_min, produksi_max = 2000, 8000
    
    # Fuzzifikasi
    turun_p = turun(permintaan, 1000, 3000)
    tetap_p = segitiga(permintaan, 1000, 3000, 5000)
    naik_p = naik(permintaan, 3000, 5000)
    
    sedikit_s = turun(persediaan, 200, 400)
    sedang_s = segitiga(persediaan, 200, 500, 800)
    banyak_s = naik(persediaan, 600, 800)
    
    # Inferensi Rules
    rules = [
        # Rule 1: Permintaan Turun AND Persediaan Sedikit => Produksi Bertambah
        (min(turun_p, sedikit_s), "bertambah"),
        # Rule 2: Permintaan Turun AND Persediaan Sedang => Produksi Berkurang
        (min(turun_p, sedang_s), "berkurang"),
        # Rule 3: Permintaan Turun AND Persediaan Banyak => Produksi Berkurang
        (min(turun_p, banyak_s), "berkurang"),
        # Rule 4: Permintaan Tetap AND Persediaan Sedikit => Produksi Bertambah
        (min(tetap_p, sedikit_s), "bertambah"),
        # Rule 5: Permintaan Tetap AND Persediaan Sedang => Produksi Berkurang
        (min(tetap_p, sedang_s), "berkurang"),
        # Rule 6: Permintaan Tetap AND Persediaan Banyak => Produksi Berkurang
        (min(tetap_p, banyak_s), "berkurang"),
        # Rule 7: Permintaan Naik AND Persediaan Sedikit => Produksi Bertambah
        (min(naik_p, sedikit_s), "bertambah"),
        # Rule 8: Permintaan Naik AND Persediaan Sedang => Produksi Bertambah
        (min(naik_p, sedang_s), "bertambah"),
        # Rule 9: Permintaan Naik AND Persediaan Banyak => Produksi Berkurang
        (min(naik_p, banyak_s), "berkurang"),
    ]
    
    # Defuzzifikasi (Rata-rata Tertimbang)
    numerator, denominator = 0, 0
    for alpha, produksi in rules:
        if produksi == "berkurang":
            z = produksi_berkurang(alpha, produksi_min, produksi_max)
        else:
            z = produksi_bertambah(alpha, produksi_min, produksi_max)
        numerator += alpha * z
        denominator += alpha
    
    return numerator / denominator if denominator != 0 else 0

# Hitung Langsung
if __name__ == "__main__":
    # Nilai Input
    permintaan = 2500  # Ubah sesuai kebutuhan
    persediaan = 500   # Ubah sesuai kebutuhan
    
    # Hasil
    produksi = fuzzy_inference(permintaan, persediaan)
    print(f"Permintaan: {permintaan}, Persediaan: {persediaan} => Produksi: {produksi:.2f} kemasan/hari")
```
